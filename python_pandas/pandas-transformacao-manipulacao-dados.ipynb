### O método json_normalize() é usado para normalizar um objeto JSON que pode ter uma ou várias hierarquias. Quando o objeto é transformado para um formato de tabela com múltiplas colunas, cada coluna representa uma chave do JSON e cada linha representa um registro. É uma forma de tornar os dados mais fáceis de manipular e analisar.

```python
dados_hospedagem = pd.json_normalize(dados_hospedagem['info_moveis'])
dados_hospedagem
```

### Números em strings applat com o lambda

```python
#Focando no tratamento da coluna preços
dados_hospedagem['preco']
```

```python
dados_hospedagem['preco'] = dados_hospedagem['preco'].astype(np.float64)
dados_hospedagem.info()
```

```python
import pandas as pd
import numpy as np

# dados da firma
dados = {
    'cargo': ['Gerencia', 'Coordenação', 'Supervisão', 'Analista', 'Assistente', 'Operação', 'Auxiliar', 'Estágio', 'Assessoria', 'Consultoria'],
    'quantidade': ['2 pessoas', '1 pessoa', '3 pessoas', '4 pessoas', '5 pessoas', '3 pessoas', '2 pessoas', '1 pessoa', '1 pessoa', '1 pessoa'],
    'salario': ['$10.000 reais', '$8.000 reais', '$7.000 reais', '$5.000 reais', '$4.000 reais', '$3.500 reais', '$3.000 reais', '$1.500 reais', '$2.500 reais', '$6.000 reais']
}

# transformando o dicionário em DataFrame
df = pd.DataFrame(dados)

# df é o DataFrame com os dados da firma
df
```

```python
df['salario'] = df['salario'].apply(lambda x: x.replace('$','').replace(',','').replace('reais','').strip())#strip apaga espaços em branco antes e depois
df['salario'] = df['salario'].astype(np.float32)
df['quantidade'] = df['quantidade'].apply(lambda x: x.replace('pessoas','').replace('pessoa','').strip())#strip apaga espaços em branco antes e depois
df['quantidade'] = df['quantidade'].astype(np.float32)
print(df)
df.info()
```

### O problema do texto <tokenização>

```python
#segregar o texto em palavras, silabas, conjunto de palavras
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.lower() #convertendo em letras minúsculas
dados_hospedagem['descricao_local']
```

### Removendo caracteres com regex

```python
#Regex conjunto de caracteres para pesquisas no texto
dados_hospedagem['descricao_local'][3169]
```

```python
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados_hospedagem['descricao_local']
```

```python
#removendo hífens soltos
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
dados_hospedagem
#verificação de caratere antes do hífem
```

### Tokenização de strings

```python
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.split()#vai pegar cada espaço em branco e treanformar em elementos de uma lista
```

```python
dados_hospedagem['comodidades'] #removendo as chaves e as aspas simples
```

```python
dados_hospedagem['comodidades'] = dados_hospedagem['comodidades'].str.replace('\{|}|\"','',regex=True).str.split()
dados_hospedagem['comodidades']
```
